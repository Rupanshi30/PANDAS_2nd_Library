# apply()

```python

import pandas as pd
# Create dataframe
data = pd.DataFrame({'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]})
data
```

### task1: add the column name: company length

```python
l=[]
for i in data["Company"]:
    l.append(len(i))
print(l)
```

```python
data["company_len"]=pd.Series(l)
data
```

```python
data["cl"]=data["Company"].apply(lambda x:len(x))
data  ##feature engineering(extracting one col from existing col)
```

```python
# new col, person name in capital
data["cap"]=data["Person"].apply(lambda x:x.upper())
data
```

```python
import numpy as np
data["log"]=data["Sales"].apply(lambda x:np.log(x))
data
```

```python
data["Person"]=data["Person"].apply(lambda x:"charly" if x=="Carl" else x)
data
```

# cut()
it is used to convert numerical to categorical variable ##concept of binning

```python
df=pd.DataFrame({'Age': np.random.randint(1, 100, 10)})
df
```

```python
df["category"]=pd.cut(x=df['Age'], bins=[1,20,40,60,80,100])    # this command creates bins accross the numbers correspondingly

df
```

```python
df["age_category"]= pd.cut(x=df['Age'], bins=[1, 20, 40, 60,80, 100],labels=["kids","Young","mid-age","Old","very-old"])   
```

# loops in the data frame

```python
info = pd.DataFrame([[1,2,3],[4,5,8],[34,56,78]],columns = ['col1','col2','col3'])  
print(info)
```

```python
for i in info:  ## gives column names
    print(i)
```

```python
for row_index,row in info.iterrows():   ## col name.iterrows()-gives row index and whole row
    print (row_index,row)  
```

```python
# we want entries from col 1 and col3 
for row_index,row in info.iterrows():   ## col name.iterrows()-gives row index and whole row
    print(info[["col1","col3"]].iloc[row_index])  
```
