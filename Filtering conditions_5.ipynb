## Value_counts(): tells frequency of each value in particular column   , efficitively work on categorical data

### group by: is used for gouping the data according to the categories and apply a function to the categories. it also helps to aggregate data efficiently

### find out the average fare spent by male and female

```python
data.groupby("sex")["fare"].mean() ## fun of pandas
```

```python
type(data.groupby("sex")["fare"].mean()) ##only one col
```

#### min,max,avg fare spent by male and female

```python
#agg for giving multiple fns
data.groupby("sex")["fare"].agg(["min","max","mean"])
```

```python
type(data.groupby("sex")["fare"].agg(["min","max","mean"]))
```

```python
x=data.groupby("sex")["fare"].agg(["min","max","mean"])
x.reset_index()
```

```python
## total fare spent by male and female
x=data.groupby("sex")["fare"].sum()
x
```

```python
## avg fare spent by each passenger class 
data.groupby([data["pclass"]])["fare"].mean()
```

```python
## find out the avg fare of male and female per each class
data.groupby(["sex","pclass"])["fare"].agg(["mean","min","max"])
```

```python
## how many people survived or not survived
data["survived"].value_counts()
```

```python
##how many male and female survived or not survived 
data.groupby("sex")["survived"].value_counts()
```

```python
## how many males and females of each class
data.groupby("sex")["pclass"].value_counts()
```

```python
## how many people survived of each class
data.groupby("pclass")["survived"].value_counts()
```

```python
## how many malesand females survived of each class
data.groupby(["sex","pclass"])["survived"].value_counts()
```

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
df

```

1. find avg and total points year wise
2. find avg,min,max points team wise and which team got max points
3. which team performed best based on avg points
4. which team performed best year wise
5. find the avg and total points of each team year wise


```python
#1
df.groupby("Year")["Points"].agg(["mean","sum"])
```

```python
#2
df1=df.groupby("Team")["Points"].agg(["mean","min","max"])
print(df1)
df1.reset_index(inplace=True)
```

```python
df1
```

```python
df1[df1["max"]==df1["max"].max()]["Team"]
```
